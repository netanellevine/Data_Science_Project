#### <b> Question 1a:
**Question:**<br>
<blockquote> approximately $\frac{1}{125}$ of the births are non identical twins and $\frac{1}{300}$ of the births are identical twins. Elvis had a twin brother who died at birth. What is the probability that Elvis was an identical twin? (we can assume the probability of boy vs girl births is identical: $\frac{1}{2}$) </blockquote>
<b> Answer: </b><br>
We shall use the Bayes' theorem to provide an answer. <br>
Bayes' theorem states:$$ P(A|B) = \frac{P(B|A) \times P(A) }{P(B)} $$
    
Let us define $A$ and $B$ for future reference: <br>
$A$: The probability that the twins are identical boys <br>
$B$: The probability for twin boys (both identical and non identical) <br>
$P(A|B)$ is the value we want to find as the answer to the question posed (given twins, what is the probability that htey are identical)<br><br>

The probability of identical boy twins is one half of the probability of identical twins (boys and girls) - using the assumption allowed in the question. let us mark this probability as $ P_{identicalBoys} $ <br>
$$ P_{identicalBoys} = \frac{1}{300} \times \frac{1}{2} = \frac{1}{600} $$ <br>
The probability of non identical boy twins is the probability of the first child being born a boy times the probability of the second child being born a boy. let us mark this probability as $ P_{nonIdenticalBoys} $ <br>
$$ P_{nonIdenticalBoys} = \frac{1}{125} \times \frac{1}{2} \times \frac{1}{2} = \frac{1}{125} \times \frac{1}{4} = \frac{1}{500} $$ <br>
$P(B)$ is the sum of the 2 possibilities of twin boys (identical and non identical)
$$ P(B) = \frac{1}{600} + \frac{1}{500} = \frac{11}{3000} = 0.00366666 $$
$P(A)$ is according to the given value of the probability for identical twins times the probability of the children to be boys ($ \frac{1}{2} $)
$$ P(A) = P_{identicalBoys} = \frac{1}{600} $$
$P(B|A)$ is the probability that there are boy twins given the boy twins are identical (slightly absurd wording, and an obvious answer - 1, because all identical twins are twins, thus all identical boy twins are boy twins)
$$ P(B|A) = 1 $$
We can build the equation according to Bayes' theorem and answer the question:
$$ P(A|B) = \frac{1 \times \frac{1}{600} }{\frac{11}{3000}} = \frac{3000}{11 \times 600} = \frac{5}{11} = 0.45454545 $$ <br>
<b>Final Answer: </b> The probability that Elvis was an identical twin is $ 0.45454545 $

<blockquote> There are 2 jars of cookies. The first jar contains 10 almond cookies and 30 chocolate ones, the second: 20 almond cookies and 20 chocolate ones. <b> Answer: </b><br>

Bayes' theorem states:$$ P(A|B) = \frac{P(B|A) \times P(A) }{P(B)} $$
    
Let us define $A$ and $B$ for future reference: <br>
$P(A)$: The probability to choose the first jar <br>
$P(B)$: The probability to choose a chocolate cookie (in general) <br>

$P(B)$: The probability for choosing a chocolate cookie in general is the sum of probabilities of either choosing the first jar and a chocolate cookie from within or choosing the second jar and a chocolate cookie from within:
$$P(B) = (\frac{1}{2} \times \frac{30}{40}) + (\frac{1}{2} \times \frac{20}{40}) = \frac{30}{80} + \frac{20}{80} = \frac{50}{80} = \frac{5}{8} = 0.625 $$
$P(A)$: The probability for choosing the first jar:
$$P(A) = \frac{1}{2}$$
$P(B|A)$: The probability of choosing a chocolate cookie given the first jar was chosen. We will use Kolmogorov's definition (from which the Bayes' theorem originates) in order to calculate this. Kolmogorov's definition states:
$$ P(B|A) = \frac{P(A) \cap P(B)}{P(A)} $$
<br>
$ P(A) \cap P(B)$: The probability of choosing the A and B, meaning choosing the first jar out of the two, and then choosing a chocolate cookie from within:
$$ P(A) \cap P(B) = \frac{1}{2} \times \frac{3}{4} = \frac{4}{6} = \frac{3}{8} = 0.375 $$
$$ P(B|A) = \frac{\frac{3}{8}}{\frac{1}{2}} = \frac{6}{8} = \frac{3}{4} = 0.75 $$
We can build the equation according to Bayes' theorem and answer the question: <br>
$$ P(A|B) = \frac{\frac{3}{4} \times \frac{1}{2} }{\frac{5}{8}} = \frac{\frac{3}{8}}{\frac{5}{8}} = \frac{3}{5} = 0.6 $$ <br>
<b>Final Answer: </b> The probability that Arik chose the first jar when it is known that the cookie chosen was chocolate is 0.6

<b> Question: </b> <blockquote> Roi is playing a dice game with Yael. Roi will roll 2 six-sided dice, and if the sum of the dice is divisible by 3, he will win 6 dollars. If the sum is not divisible by 3, he will lose 3 dollars. What is Roiâ€™s expected value of playing this game?</blockquote>
<b> Answer: </b><br>
Let's consider all options for the dice to fall (noticing that (for example) the option of the first die falling: 1, and the second: 2, is not the same as the opposite: first die falling: 2 and the second: 1. <br>

<1, 4> <br>
<1, 5> <br>

To calculate the expected average profit/loss from such a game, we have to add the 2 probabilities multiplied by the winnings/losings in each option.
Lets write all available information:
$P_{win} = \frac{1}{3}$<br>
$P_{lose} = \frac{2}{3}$<br>
$P_{+atWin} = 6$ dollars <br>
$P_{+atLoss} = -3$ dollars <br><br>

#### <b> Question 4:
<b> Question: </b> <blockquote>Different dealers may sell the same car for different prices. The sale prices for a particular car are normally distributed with a mean and standard deviation of 26,000 dollars and 2,000 dollars, respectively. Suppose we select one of these cars at random. Let X = the sale price (in thousands of dollars) for the selected car. Find P(26<X<30)
</blockquote>
<b> Answer: </b><br>
If we were to imagine a normal distribution graph and its bell like shape, the top point of the graph would be at (0,26000). When the standard deviation is also known, we can surmise that (looking at the x axis) from 0 to $\sigma$ fall 34.1% of randomly chosen values, where $\sigma$ is the value of the standard deviation. Likewise, from $\sigma$ to $2\sigma$ fall 13.6% of randomly chosen values.<br>
The probability we were asked to provide in the question is the probability of randomly choosing a car and falling from 26000 to 30000 (note that the difference between the 2 values is 4000 which is exactly twice the standard deviation, meaning $2\sigma$).<br>
All that left is to add the two percentages and transfer them to probability:
$$ P(26000 < X < 30000) = 34.1\% + 13.6\% = 47.7\% = 0.477 $$<br>
<b>Final Answer:</b> $ P(26000 < X < 30000) = 0.477 $

#### <b> Question 6:
<b> Question: </b> <blockquote>A company has 500 employees, and 60% of them have children. Suppose that we randomly select 4 of these employees. What is the probability that exactly 3 of the 4 employees selected have children?</blockquote>
<b> Answer: </b><br>
Let us first determine the number of employees with children and the amount of those without: 
$$ 500 \times 0.6 = 300 $$
$$ 500 - 300 = 200 $$
From the calculations follows that there are 300 employees with children and 200 without. Next, we have to calculate the probability that exactly 3 of the 4 randomly chosen have children (note that when one person is chosen, the amount of unchosen people lessens by 1 and the group from which the person was chosen lessens by 1 too):
$$ \frac{300}{500} \times \frac{299}{499} \times \frac{298}{498} \times \frac{200}{497} = 0.08657 $$<br>
What remains is multiplying this answer by 4, due to there being 4 possible combinations whether the employee who has no children is chosen first, second, third or fourth. Each of these possibilities has the same probability of occurring, hence we do the following:
$$ 0.08657 \times 4 = 0.34628 $$<br>
<b>Final Answer:</b> The probability that exactly 3 of the 4 employees randomly chosen have children is 0.34628
